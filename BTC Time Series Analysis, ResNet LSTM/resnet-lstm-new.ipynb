```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load data
btc_data = pd.read_csv('/kaggle/input/btc-usd-maxcsv/btc-usd-max.csv',
                       parse_dates=['snapped_at'], index_col='snapped_at')
if not isinstance(btc_data.index, pd.DatetimeIndex):
    btc_data.index = pd.to_datetime(btc_data.index, errors='coerce')
btc_data = btc_data.iloc[250:]

plt.figure(figsize=(14, 7))
plt.plot(btc_data['price'], label='Bitcoin')
plt.title('Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()

# Create sequences
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - 2 * seq_length):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length:i + 2 * seq_length] - data[i + seq_length - 1:i + 2 * seq_length - 1]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Normalize data and split for training/trading
sequence_length = 6
scaler_price = MinMaxScaler()
btc_price = scaler_price.fit_transform(btc_data['price'].values.reshape(-1, 1))
scaler_volume = MinMaxScaler()
btc_volume = scaler_volume.fit_transform(btc_data['total_volume'].values.reshape(-1, 1))
btc_data_normalized = np.hstack((btc_price, btc_volume))

cut_off_date = btc_data.index.max() - pd.DateOffset(months=2)
mask_train = btc_data.index <= cut_off_date
mask_last2 = btc_data.index > cut_off_date
btc_data_normalized_train = btc_data_normalized[mask_train]
btc_data_df_last_two_months = btc_data.loc[mask_last2].copy()

X_seq, y_seq = create_sequences(btc_data_normalized_train, sequence_length)
X_train_full, X_test, y_train_full, y_test = train_test_split(X_seq, y_seq, test_size=0.2, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)

# Residual LSTM model
class ResidualWrapper(tf.keras.Model):
    def __init__(self, model):
        super().__init__()
        self.model = model
    def call(self, inputs, *args, **kwargs):
        delta = self.model(inputs, *args, **kwargs)
        baseline = inputs[:, -1:, :]
        output = baseline + tf.cumsum(delta, axis=1)
        return output

lstm_model = tf.keras.Sequential([
    tf.keras.layers.LSTM(128, return_sequences=True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(128, return_sequences=True),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(2, kernel_initializer=tf.initializers.zeros()))
])
residual_lstm = ResidualWrapper(lstm_model)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
residual_lstm.compile(optimizer=optimizer, loss='mse')

history = residual_lstm.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=128,
    epochs=60,
    verbose=1,
)

plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

y_pred = residual_lstm.predict(X_test)
y_pred = y_pred[:, -sequence_length:, :]
y_test_flat = y_test.flatten()
y_pred_flat = y_pred.flatten()
print("Evaluation Metrics on Test Set:")
print("MAE:", mean_absolute_error(y_test_flat, y_pred_flat))
print("MSE:", mean_squared_error(y_test_flat, y_pred_flat))
print("RMSE:", np.sqrt(mean_squared_error(y_test_flat, y_pred_flat)))

# Predictions on last two months
btc_price_last2 = scaler_price.transform(btc_data_df_last_two_months['price'].values.reshape(-1, 1))
btc_volume_last2 = scaler_volume.transform(btc_data_df_last_two_months['total_volume'].values.reshape(-1, 1))
btc_data_last2_normalized = np.hstack((btc_price_last2, btc_volume_last2))
btc_sequences_last2, _ = create_sequences(btc_data_last2_normalized, sequence_length)

y_pred_rescaled_all = []
for i in range(len(btc_sequences_last2)):
    seq = np.expand_dims(btc_sequences_last2[i], axis=0)
    y_pred_seq = residual_lstm.predict(seq)
    y_pred_price = y_pred_seq[0, :, 0].reshape(-1, 1)
    y_pred_rescaled = scaler_price.inverse_transform(y_pred_price)
    y_pred_rescaled_all.append(y_pred_rescaled.flatten())

# Ichimoku calculation (using Tenkan, Kijun, Senkou spans)
def compute_ichimoku(df, price_col='price'):
    period9, period26, period52 = 9, 26, 52
    df['tenkan_sen'] = (df[price_col].rolling(window=period9).max() + df[price_col].rolling(window=period9).min()) / 2
    df['kijun_sen'] = (df[price_col].rolling(window=period26).max() + df[price_col].rolling(window=period26).min()) / 2
    df['senkou_a'] = ((df['tenkan_sen'] + df['kijun_sen']) / 2).shift(period26)
    df['senkou_b'] = ((df[price_col].rolling(window=period52).max() + df[price_col].rolling(window=period52).min()) / 2).shift(period26)
    return df

ichimoku_df = compute_ichimoku(btc_data_df_last_two_months.copy())

# New strategy: combine model predictions, Ichimoku (using kijun_sen), and anti-martingale money management
initial_fraction = 0.05
current_fraction = initial_fraction
capital = 10000
position = 0
entry_price = 0
profit_history = []

for i in range(len(y_pred_rescaled_all)):
    current_date = btc_data_df_last_two_months.index[i]
    current_price = btc_data_df_last_two_months.loc[current_date, 'price']
    pred_seq = y_pred_rescaled_all[i]
    pred_trend = pred_seq[-1] - pred_seq[0]
    threshold = current_price * 0.005
    if pred_trend > threshold:
        model_signal = 1
    elif pred_trend < -threshold:
        model_signal = -1
    else:
        model_signal = 0

    if current_date in ichimoku_df.index:
        kijun = ichimoku_df.loc[current_date, 'kijun_sen']
    else:
        continue
    if current_price > kijun:
        ichimoku_signal = 1
    elif current_price < kijun:
        ichimoku_signal = -1
    else:
        ichimoku_signal = 0

    signal = 0
    if model_signal == 1 and ichimoku_signal == 1:
        signal = 1
    elif model_signal == -1 and ichimoku_signal == -1:
        signal = -1

    if signal == 1 and position == 0:
        invest = current_fraction * capital
        position = invest / current_price
        capital -= invest
        entry_price = current_price
        print(f"{current_date.date()} BUY: Invest {invest:.2f}, Pos {position:.4f}")
    elif signal == -1 and position > 0:
        proceeds = position * current_price
        profit = proceeds - (position * entry_price)
        capital += proceeds
        print(f"{current_date.date()} SELL: Proceeds {proceeds:.2f}, Profit {profit:.2f}")
        if profit > 0:
            current_fraction = min(current_fraction * 1.5, 0.5)
        else:
            current_fraction = initial_fraction
        profit_history.append(profit)
        position = 0
        entry_price = 0

if position > 0:
    last_date = btc_data_df_last_two_months.index[-1]
    last_price = btc_data_df_last_two_months.loc[last_date, 'price']
    proceeds = position * last_price
    profit = proceeds - (position * entry_price)
    capital += proceeds
    print(f"{last_date.date()} FINAL SELL: Proceeds {proceeds:.2f}, Profit {profit:.2f}")
    profit_history.append(profit)
    position = 0

print(f"Final Capital: {capital:.2f}")
```
